```python
import glob
import subprocess

for f in glob.glob("data/recordings/*.m4a"):
    result = subprocess.run([
        "whisper", f,
        "--language", "pt",
        "--task", "transcribe",
        "--model", "medium",
        "--output_format", "srt",
        "--output_dir", "data/transcriptions"
    ], capture_output=True, text=True)
```
